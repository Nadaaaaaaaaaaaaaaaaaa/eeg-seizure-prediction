```python
# === Fonction pour parser un fichier .seizures ===
def parse_seizure_file(filepath):
    entries = []
    with open(filepath, 'r') as file:
        lines = file.readlines()
        times = [int(line.strip()) for line in lines if line.strip().isdigit()]
        edf_filename = os.path.basename(filepath).replace(".edf.seizures", ".edf")

        for i in range(0, len(times), 2):
            entries.append({
                "edf_file": edf_filename,
                "seizure_start_time": times[i],
                "seizure_end_time": times[i+1],
                "duration": times[i+1] - times[i],
                "has_seizure": True,
                "source": "seizures_file"
            })
    return entries
```

```python
def parse_summary_file(filepath):
    entries = []
    with open(filepath, 'r') as file:
        lines = file.readlines()
        current_file = None
        i = 0
        while i < len(lines):
            line = lines[i]
            match_file = re.search(r'File Name:\s*(chb\d+_\d+\.edf)', line)
            if match_file:
                current_file = match_file.group(1)

            match_seizure = re.search(r'Seizure Start Time:\s*(\d+)', line)
            if match_seizure and current_file:
                start_time = int(match_seizure.group(1))
                # Vérifie s’il y a une ligne suivante
                if i + 1 < len(lines):
                    end_line = lines[i + 1]
                    match_end = re.search(r'Seizure End Time:\s*(\d+)', end_line)
                    if match_end:
                        end_time = int(match_end.group(1))
                        entries.append({
                            "edf_file": current_file,
                            "seizure_start_time": start_time,
                            "seizure_end_time": end_time,
                            "duration": end_time - start_time,
                            "has_seizure": True,
                            "source": "summary"
                        })
                    i += 1  # On avance d'une ligne supplémentaire pour ne pas relire `end_line`
            i += 1  # Ligne suivante
    return entries
```

```python
# === 1. Traitement des fichiers .seizures ===
for filename in os.listdir(chb_folder):
    if filename.endswith(".edf.seizures"):
        full_path = os.path.join(chb_folder, filename)
        annotations.extend(parse_seizure_file(full_path))

# === 2. Traitement du fichier summary ===
summary_path = os.path.join(chb_folder, "chb01-summary.txt")
if os.path.exists(summary_path):
    annotations.extend(parse_summary_file(summary_path))
```

### # Appliquer un filtre Butterworth passe-bande 

```python
# === 1. Filtrage Butterworth ===
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a
```

```python
def apply_filter(signal, lowcut, highcut, fs, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order)
    return filtfilt(b, a, signal)
```

```python
# === 2. Traitement de chaque fichier EDF ===
def process_edf_file(filepath, output_folder, lowcut=0.5, highcut=40.0):
    f = pyedflib.EdfReader(filepath)
    n_signals = f.signals_in_file
    fs = f.getSampleFrequency(0)
    signal_labels = f.getSignalLabels()

    print(f"Traitement : {os.path.basename(filepath)}")
    all_filtered = []

    for i in range(n_signals):
        raw = f.readSignal(i)
        filtered = apply_filter(raw, lowcut, highcut, fs)
        all_filtered.append(filtered)

    f._close()
    del f

    filtered_array = np.array(all_filtered)  # shape: (channels, length)
    df = pd.DataFrame(filtered_array.T, columns=signal_labels)

    filename_csv = os.path.splitext(os.path.basename(filepath))[0] + "_filtered.csv"
    output_path = os.path.join(output_folder, filename_csv)
    df.to_csv(output_path, index=False)

    return df, fs  # pour affichage
```

```python
# === 3. Affichage d’un extrait pour vérification ===
def plot_segment(df, fs, channel=0, seconds=10):
    samples = int(fs * seconds)
    segment = df.iloc[:samples, channel]
    plt.figure(figsize=(12, 4))
    plt.plot(segment)
    plt.title(f"Signal filtré - Canal {df.columns[channel]} ({seconds} sec)")
    plt.xlabel("Échantillons")
    plt.ylabel("Amplitude (µV)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### # Découper les signaux filtrés (déjà enregistrés en .csv) en segments de 2 secondes

```python
def split_csv_into_segments(csv_path, fs, segment_duration=2, output_dir="segments"):
    os.makedirs(output_dir, exist_ok=True)
    
    df = pd.read_csv(csv_path)
    total_samples = len(df)
    segment_samples = int(fs * segment_duration)
    num_segments = total_samples // segment_samples

    base_name = os.path.splitext(os.path.basename(csv_path))[0]

    for i in range(num_segments):
        start = i * segment_samples
        end = start + segment_samples
        segment_df = df.iloc[start:end]
        
        segment_filename = f"{base_name}_segment_{i:03}.csv"
        segment_path = os.path.join(output_dir, segment_filename)
        segment_df.to_csv(segment_path, index=False)
```

```python
# === Paramètres ===
SEGMENTS_FOLDER = 'chb01/segments'
LABELED_FOLDER = 'labeled_segments'
ANNOTATION_FILE = 'chb01/annotations.csv'
SEGMENT_DURATION = 2  # en secondes

os.makedirs(LABELED_FOLDER, exist_ok=True)

# === Charger les annotations ===
annotations_df = pd.read_csv(ANNOTATION_FILE)
```

```python
seizure_dict = {}

for _, row in annotations_df.iterrows():
    if row['has_seizure'] == True or str(row['has_seizure']).lower() == 'true':
        fname = row['edf_file']
        start = row['seizure_start_time']
        end = row['seizure_end_time']
        if pd.notna(start) and pd.notna(end):
            if fname not in seizure_dict:
                seizure_dict[fname] = []
            seizure_dict[fname].append((start, end))
```

```python
import pprint
pprint.pprint(seizure_dict)
```

```python
# === Fonction d'étiquetage ===
def label_segment(segment_start, segment_end, seizures):
    for start, end in seizures:
        preictal_start = max(0, start - 30 * 60)
        postictal_end = end + 10 * 60

        if segment_start >= start and segment_end <= end:
            return 'ictal'
        elif segment_end > preictal_start and segment_start < start:
            return 'preictal'
        elif segment_start < postictal_end and segment_end > end:
            return 'postictal'
    return 'interictal'
```

```python
labels_summary = []

for filename in os.listdir(SEGMENTS_FOLDER):
    if filename.endswith('.csv'):
        full_path = os.path.join(SEGMENTS_FOLDER, filename)
        signal = pd.read_csv(full_path)

        # Extraire le nom de base original (sans "_filtered" et sans "_segment")
        raw_base = filename.split('_segment_')[0]
        if raw_base.endswith('_filtered'):
            edf_base = raw_base.replace('_filtered', '') + '.edf'
        else:
            edf_base = raw_base + '.edf'

        # Index du segment
        segment_index = int(filename.split('_segment_')[1].split('.csv')[0])

        # Calcul des temps
        segment_start = segment_index * SEGMENT_DURATION
        segment_end = segment_start + SEGMENT_DURATION

        seizures = seizure_dict.get(edf_base, [])
        label = label_segment(segment_start, segment_end, seizures)

        # Enregistrer avec étiquette dans le nom
        labeled_filename = f"{raw_base}_segment_{segment_index}_{label}.csv"
        labeled_path = os.path.join(LABELED_FOLDER, labeled_filename)
        signal.to_csv(labeled_path, index=False)

        # Ajouter à la table de résumé
        labels_summary.append({
            'original_file': edf_base,
            'segment_index': segment_index,
            'start_time': segment_start,
            'end_time': segment_end,
            'label': label,
            'path': labeled_path
        })
```

```python
# === Sauvegarde du fichier CSV de résumé ===
labels_df = pd.DataFrame(labels_summary)
labels_df.to_csv('segments_labels.csv', index=False)

print(f"✅ Segments labellisés et enregistrés dans : {LABELED_FOLDER}")
print(f"📄 Résumé des étiquettes : segments_labels.csv")
```

```python
segment_folder = 'labeled_segments'  # Dossier des segments CSV
output_file = 'features_segments.csv'  # Fichier de sortie
```

```python
# === Bandes de fréquences EEG ===
bands = {
    'delta': (0.5, 4),
    'theta': (4, 8),
    'alpha': (8, 13),
    'beta': (13, 30),
    'gamma': (30, 100)
}
```

```python
def extract_band_power(data, fs, band):
    fmin, fmax = band
    freqs, psd = welch(data, fs=fs, nperseg=len(data))
    band_power = np.trapz(psd[(freqs >= fmin) & (freqs <= fmax)], freqs[(freqs >= fmin) & (freqs <= fmax)])
    return band_power
```

```python
def extract_features(segment_path):
    df = pd.read_csv(segment_path)
    features = {}
    for col in df.columns:
        signal = df[col].values

        # === Features temporelles ===
        features[f'{col}_mean'] = np.mean(signal)
        features[f'{col}_std'] = np.std(signal)
        features[f'{col}_var'] = np.var(signal)
        features[f'{col}_min'] = np.min(signal)
        features[f'{col}_max'] = np.max(signal)
        features[f'{col}_skew'] = skew(signal)
        features[f'{col}_kurtosis'] = kurtosis(signal)

        # === Entropie de Shannon ===
        hist, _ = np.histogram(signal, bins=100, density=True)
        hist = hist[hist > 0]
        features[f'{col}_shannon'] = entropy(hist)

        # === Puissance dans chaque bande ===
        for band_name, band_range in bands.items():
            power = extract_band_power(signal, fs, band_range)
            features[f'{col}_power_{band_name}'] = power

    # Ajout du label depuis le nom du fichier
    label = os.path.basename(segment_path).split('_')[-1].replace('.csv', '')
    features['label'] = label

    return features
```

```python
# === Traitement de tous les segments ===
all_features = []
for filename in os.listdir(segment_folder):
    if filename.endswith('.csv'):
        path = os.path.join(segment_folder, filename)
        feats = extract_features(path)
        feats['segment'] = filename
        all_features.append(feats)
```
